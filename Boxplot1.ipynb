```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   # ✅ add this line

data = {"salary": [25, 28, 30, 32, 29, 27, 500]}
df = pd.DataFrame(data)

# Create boxplot
sns.boxplot(x=df['salary'])

plt.show()
```
